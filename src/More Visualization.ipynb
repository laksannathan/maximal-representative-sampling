```python
from pathlib import Path
import os
from PIL import Image 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


path = Path(os.getcwd()).parent

gbs = pd.read_csv(os.path.join(path, 'data/gbs_processed.csv'), encoding = "ISO-8859-1", delimiter = ',')
gesis = pd.read_csv(os.path.join(path, 'data/gesis_processed.csv'), encoding = "ISO-8859-1", delimiter = ',')
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(12,12))

sns.heatmap(gbs.iloc[:,0:14].isnull(), cbar=False, ax=axs[0])
sns.heatmap(gbs.iloc[:,14:41].isnull(), cbar=False, ax=axs[1])

plt.savefig(os.path.join(path, 'tex/fig/gbs_missing.png'))
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(12,12))

sns.heatmap(gesis.iloc[:,0:14].isnull(), cbar=False, ax=axs[0])
sns.heatmap(gesis.iloc[:,14:41].isnull(), cbar=False, ax=axs[1])

plt.savefig(os.path.join(path, 'tex/fig/gesis_missing.png'))
```

```python
fig = plt.figure(figsize = (15,20))
ax = fig.gca()

gbs[['Schlechter Schlaf', 'Leben genießen',
       'Zu Nichts aufraffen', 'Alles anstrengend',
       'Zufriedenheit Leben', 'Aktiv',
       'Verärgert', 'Wach', 'Ängstlich',
       'Nervoes', 'Druck', 'Optimismus Zukunft']].hist(ax=ax)

plt.savefig(os.path.join(path, 'tex/fig/gbs_hist.png'))
plt.show()
```

```python
image_file = Image.open(os.path.join(path, 'tex/fig/gbs_hist.png'))
image_file = image_file.convert('1') # convert image to black and white
image_file.save(os.path.join(path, 'tex/fig/gbs_hist.png'))
```

```python
fig = plt.figure(figsize = (15,20))
ax = fig.gca()

gesis[['Schlechter Schlaf', 'Leben genießen',
       'Zu Nichts aufraffen', 'Alles anstrengend',
       'Zufriedenheit Leben', 'Aktiv',
       'Verärgert', 'Wach', 'Ängstlich',
       'Nervoes', 'Druck', 'Optimismus Zukunft']].hist(ax=ax)

plt.savefig(os.path.join(path, 'tex/fig/gesis_hist.png'))
plt.show()
```
