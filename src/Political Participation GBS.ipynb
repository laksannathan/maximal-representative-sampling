### This notebook trains classifier to predict political participation given GBS.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

gbs = pd.read_csv('gbs_processed.csv', encoding = "ISO-8859-1", delimiter = ',')
gbs= gbs.fillna(gbs.median())
gbs.head()
```

```python
y = gbs.Wahlteilnahme
X = gbs.drop(columns = ['GBS-CODE', 'Gruppe', 'Wahlteilnahme','Desinteresse Politiker', 'Zufriedenheit Leben'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X.head(300), y[:300], test_size=0.33, random_state=42)
```

```python
from sklearn import tree
from sklearn.metrics import mean_squared_error

clf = tree.DecisionTreeRegressor(max_depth=3)
clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print(mean_squared_error(y_test, y_pred))

clf.feature_importances_
```
