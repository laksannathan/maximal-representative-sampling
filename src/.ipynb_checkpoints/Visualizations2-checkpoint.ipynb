```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi

fs1a = ['v2673_1_bfi_bfi1','v2673_1_bfi_bfi2','v2673_1_bfi_bfi3','v2673_1_bfi_bfi4','v2673_1_bfi_bfi5',
        'v2673_1_bfi_bfi6','v2673_1_bfi_bfi7','v2673_1_bfi_bfi8','v2673_1_bfi_bfi9','v2673_1_bfi_bfi10']

fs1b = ['bdze001a','bdze002a','bdze003a','bdze004a','bdze005a','bdze006a','bdze007a','bdze008a','bdze009a','bdze010a']

name = ['Zurueckhaltend','leicht Vertrauen','Faulheit','Entspannt','wenig kuenstlerisches Interesse','Gesellig',
        'Andere kritisieren','Gruendlich','Nervoes','Phantasievoll']

gbs = pd.read_csv('gbs_bfi_lev.csv', usecols=fs1a)
gesis = pd.read_csv('ZA5665_a1_ba-bf_v22-0-0.dta.csv',engine='python',usecols=fs1b)

gbs['Umfrage'] = 'GBS'
gesis['Umfrage'] = 'GESIS'

gbs.rename(index=str, columns={k:v for k,v in zip(fs1a, name)}, inplace=True)
gesis.rename(index=str, columns={k:v for k,v in zip(fs1b, name)}, inplace=True)

df = pd.concat([gbs,gesis])
df.head(5)
```

```python
mapX = {'5 Trifft voll und ganz zu':5,
        '4 Eher zutreffend':4,
        '3 Weder noch':3,
        '2 Trifft eher nicht zu':2,
        '1 Trifft überhaupt nicht zu':1,
        
        'trifft voll und ganz zu (5)':5,
        'eher zutreffend (4)':4,
        'weder noch (3)':3,     
        'trifft eher nicht zu (2)':2,
        'trifft ueberhaupt nicht zu (1)':1, 
        
        'Unit nonresponse':np.nan,
        'Not in panel':np.nan,
        'Item nonresponse':np.nan,
        'Not reached':np.nan,
        'Ambiguous answer':np.nan,
        'keine Angabe':np.nan}

df = df.replace(mapX)
df = df.dropna()
#df.to_csv('preprocessed1.csv', sep=',', encoding='utf-8', index=False)

df.head(5)
```

```python
gesis = gesis.replace(mapX)
gbs = gbs.replace(mapX)
gbs = gbs.dropna()
gesis = gesis.dropna()

def normalizeFeatures(frame, att, n):
    x = [0,0,0,0,0]
    for i in frame[att]:
        x[int(i)-1] = x[int(i)-1] + 1
    x = [i/n for i in x]
    return x

def plotDist(a,b, name):
    print('a:',a)
    fig, ax = plt.subplots()
    index = np.arange(5)
    bar_width = 0.4

    rects1 = plt.bar(index, a, bar_width, alpha=0.8, color='grey', label='Gesis')
    rects2 = plt.bar(index + bar_width, b, bar_width, alpha=0.8, color='black', label='GBS')

    plt.xticks(index + bar_width - 0.2, ['1','2','3','4','5'])
    plt.legend()
    plt.tight_layout()
    #plt.savefig(name)
    plt.show()

for att in name:
    plotDist(normalizeFeatures(gesis, att, len(gesis['Umfrage'])),
             normalizeFeatures(gbs, att, len(gbs['Umfrage'])),att)
```

```python
#import math

for i in name:
    left = normalizeFeatures(gesis, i, len(gesis['Umfrage']))
    right = normalizeFeatures(gbs, i, len(gbs['Umfrage']))

    most = max(max(left),max(right))

    # Set data
    pdf = pd.DataFrame([left, right], columns=['1','2','3','4','5'])

    # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(5) * 2 * pi for n in range(5)]
    angles += angles[:1]

    # Initialise the spider plot
    ax = plt.subplot(111, polar=True)

    # If you want the first axis to be on top:
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)

    # Draw one axe per variable + add labels yet
    plt.xticks(angles[:-1], ['1','2','3','4','5'])

    # Draw ylabels
    ax.set_rlabel_position(0)
    plt.yticks([most*1/4,most*2/4,most*3/4], [round(most*1/4,2),round(most*2/4,2),round(most*3/4,2)], color="grey", size=7)
    plt.ylim(0,most)

    # GESIS
    values=pdf.loc[0].values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, color='black', linewidth=1, linestyle='solid', label="GESIS")
    ax.fill(angles, values, 'grey', alpha=0.2)

    # GBS
    values=pdf.loc[1].values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, color='black', linewidth=1, linestyle='dashed', label="GBS")
    ax.fill(angles, values, 'black', alpha=0.1)

    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    plt.xlabel(i)
    plt.savefig(i)
    plt.show()
```

```python
plt.errorbar(range(0,10), gesis.mean(), gesis.std()**2, linestyle='None', marker='.', color='black')
plt.xticks(np.arange(10), ['Zur','Ver','Fau','Ent','Kue','Ges','Kri','Gru','Ner','Pha'])
plt.savefig('std_gesis')
plt.show()
```
